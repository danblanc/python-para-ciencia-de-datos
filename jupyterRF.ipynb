```python
import pandas as pd
import numpy as np

# Definir los nombres de las columnas
Variables= ['cod_reg', 'cod_dep', 'cod_loc', 'pad', 'block', 'EP', 'uni', 'sup_predio','sup_edificada', 'V_cat_terreno',
            'V_cat_mejoras',  'V_cat_total', 'V_impuestos', 'Fecha_UDJ', 'Vigencia_UDJ']  

# Especificar los tipos de datos para las columnas
dtypes = {
    'cod_reg': str,
    'cod_dep': str,
    'cod_loc': str,  # O str, dependiendo de tus datos
    'pad': int,
    'block': str,
    'EP': str,
    'uni': int,
    'sup_predio': int,
    'sup_edificada': int,
    'V_cat_terreno': int,
    'V_cat_mejoras': int,
    'V_cat_total': int,
    'V_impuestos': int,
    'Fecha_UDJ': bool,
    'Vigencia_UDJ': bool
}

# Leer el archivo CSV y asignar los nombres a las columnas
P_Urbanos = pd.read_csv('C:/Users/Nico/Desktop/archivos proyecto PYTHON/Padrones Urbanos.csv', sep=',', encoding='cp1252', header=None, names=Variables)

# Mostrar las primeras filas del archivo
P_Urbanos
```

```python
# Filtrar filas donde 'cod_dep' es igual a 'V' que es Montevideo
filtro_dep = P_Urbanos[P_Urbanos['cod_dep'] == 'V']

# Mostrar el DataFrame filtrado
filtro_dep
```

```python
#CUantas filas tiene el archivo
filtro_dep[[ 'dia','mes','anio']] = filtro_dep['Fecha_UDJ'].str.split('/', expand=True)

filtro_dep
```

```python
#CUantas filas tiene el archivo
num_filas = filtro_dep.shape[0]

print(num_filas)
```

```python
# Ver los tipos de datos de cada columna
tipos_datos = filtro_dep.dtypes

# Mostrar los tipos de datos
print(tipos_datos)
```

```python
# Reemplazar las cadenas vacÃ­as con 0 en las columnas 'dia', 'mes' y 'anio'
columnas_a_reemplazar = ['dia', 'mes', 'anio']
filtro_dep[columnas_a_reemplazar] = filtro_dep[columnas_a_reemplazar].replace('', '0')


filtro_dep
```

```python
# Convertir las columnas a entero
filtro_dep[columnas_a_reemplazar] = filtro_dep[columnas_a_reemplazar].astype(int)
```

```python
# Ver los tipos de datos de cada columna
tipos_datos = filtro_dep.dtypes

# Mostrar los tipos de datos
print(tipos_datos)
```

```python
# Leer el archivo CSV y asignar los nombres a las columnas
Permisos = pd.read_csv('C:/Users/Nico/Desktop/archivos proyecto PYTHON/permisos_construccion.csv', sep=';')

Permisos.rename(columns={'padron': 'pad'}, inplace=True)
# Mostrar las primeras filas del archivo
Permisos
```

```python
# Ver los tipos de datos de cada columna
tipos_datos = Permisos.dtypes

# Mostrar los tipos de datos
print(tipos_datos)
```

```python
#Crear en permisos la variable cod_dep
Permisos['cod_dep'] = 'V'
Permisos
```

```python
# Guardar el DataFrame macheado como CSV
#filtro_per.to_csv('C:/Users/Nico/Desktop/archivos proyecto PYTHON/Permisos_2022a2024.csv', index=False)
```

```python
# Machear los DataFrames usando un left join
cat_per = pd.merge(filtro_dep, Permisos, on=['anio','mes','pad'], how='left')


# Mostrar el DataFrame resultante
print(cat_per)

#Guardar el archivo macheado
cat_per.to_csv('C:/Users/Nico/Desktop/Archivos para proyecto Python/cat_per.csv', index=False)
```

```python
# por cada padron puede haber mÃ s de un permiso, por ello para este anÃ lisis elimino duplicados
df_sin_duplicados = cat_per.drop_duplicates(subset=['anio', 'mes', 'pad'])

#Montevideo tiene 510298 padrones urbanos 
#Para saber por aÃ±o cuantos padrones urbanos tuvieron permisos en un aÃ±o

resultado = df_sin_duplicados.groupby(['anio','mes'])['destino'].count().reset_index(name='conteo_destino')

# Mostrar el resultado
print(resultado)
```
