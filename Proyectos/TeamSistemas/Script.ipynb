**Análisis Interactivo de Datos con Visualización Avanzada**

**Descripción General**

Este proyecto consiste en una aplicación interactiva de análisis de datos, utilizando diferentes librerías de visualización y manipulación de datos en Python. La aplicación permite cargar y procesar conjuntos de datos, y genera gráficos interactivos y mapas dinámicos para ayudar en la toma de decisiones basada en datos.

La aplicación está construida sobre Streamlit, lo que permite a los usuarios interactuar con los datos y obtener insights visuales en tiempo real. El objetivo es ofrecer una herramienta accesible para el análisis de datos de forma intuitiva y visualmente atractiva.

**Librerías Utilizadas**
**Streamlit** (streamlit):
Utilizado para crear la interfaz de usuario interactiva. Permite el despliegue rápido de dashboards y gráficos interactivos sin necesidad de configurar un servidor web.

**Seaborn** (seaborn):
Librería basada en Matplotlib, especializada en la visualización de datos estadísticos. Facilita la creación de gráficos como distribuciones, heatmaps y gráficos de regresión.

**Pandas** (pandas):Esencial para la manipulación y análisis de datos. Se utiliza para cargar, limpiar y preparar los datos que se utilizarán en las visualizaciones.

**Matplotlib** (matplotlib):
Librería de trazado 2D que proporciona control total sobre los gráficos. Se utiliza junto con Seaborn para crear visualizaciones más detalladas.

**Numpy** (numpy):
Biblioteca para trabajar con arreglos multidimensionales y realizar cálculos matemáticos de manera eficiente. Se utiliza principalmente para operaciones numéricas en los datos.

**Plotly Express** (plotly.express):
Proporciona una API simple para la creación de gráficos interactivos. Es ideal para gráficos como scatter plots, line plots, histograms, etc., con soporte para interactividad.

**Plotly Graph Objects** (plotly.graph_objects):
Utilizado para personalizar gráficos interactivos avanzados y construir gráficos 3D, mapas coropléticos y gráficos combinados.

**Folium** (folium):
Permite la creación de mapas interactivos usando Leaflet.js. Utilizado para visualizar datos geoespaciales y trazar ubicaciones geográficas en un mapa.

**Streamlit-Folium** (streamlit-folium):
Un complemento para integrar mapas creados con Folium dentro de las aplicaciones de Streamlit, lo que permite agregar interactividad geoespacial a la app.

**Altair** (altair):
Librería declarativa para la creación de gráficos interactivos. Utilizada para crear gráficos de alta calidad y visualizaciones estadísticas a través de una API simple y declarativa.

**Matplotlib Ticker, Colorbar, Colors** (matplotlib.ticker, matplotlib.colorbar, matplotlib.colors):
Modifican el formato y la estética de los gráficos en Matplotlib, como la escala de colores, las barras de colores y las etiquetas en los ejes.
